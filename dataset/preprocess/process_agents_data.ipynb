```python
import os
import json

folder_path = "/Users/vaibhavmishra/Desktop/btx-game-aicode/clash_squad_agent_partition"
json_files = [f for f in os.listdir(folder_path) if f.endswith('.json')]

weapon_list = []

def prepare_features(agent_time_quanta):
    # for key in agent_time_quanta.keys():
    #     print(key)
    feature_list = []
    feature_list.append(agent_time_quanta['agentPosition']['x'] /1000) #1
    feature_list.append(agent_time_quanta['agentPosition']['y'] /1000) #2
    feature_list.append(agent_time_quanta['agentPosition']['z'] /1000) #3
    feature_list.append(agent_time_quanta['agentRotation']['y'] /360) #4
    feature_list.append(agent_time_quanta['agentForward']['x'] ) #5
    feature_list.append(agent_time_quanta['agentForward']['y']) #6
    feature_list.append(agent_time_quanta['agentForward']['z']) #7
    feature_list.append(agent_time_quanta['health']/100) #8
    
    if agent_time_quanta['weapon'] not in weapon_list:
        weapon_list.append(agent_time_quanta['weapon'])
    feature_list.append(weapon_list.index(agent_time_quanta['weapon'])) #9

    feature_list.append(1 if agent_time_quanta['islos'] else 0) #10
    

    feature_list.append(agent_time_quanta['targetPosition']['x'] /1000) #1
    feature_list.append(agent_time_quanta['targetPosition']['y'] /1000) #2    
    feature_list.append(agent_time_quanta['targetPosition']['z'] /1000) #3
    feature_list.append(agent_time_quanta['targetRotation']['y'] /360) #4
    feature_list.append(agent_time_quanta['targetForward']['x']) #5
    feature_list.append(agent_time_quanta['targetForward']['y']) #6
    feature_list.append(agent_time_quanta['targetForward']['z']) #7
    feature_list.append(agent_time_quanta['directionToTarget']['x']) #8
    feature_list.append(agent_time_quanta['directionToTarget']['y']) #9
    feature_list.append(agent_time_quanta['directionToTarget']['z']) #10
    feature_list.append(agent_time_quanta['cross']['x']) #11
    feature_list.append(agent_time_quanta['cross']['y']) #12
    feature_list.append(agent_time_quanta['cross']['z']) #13
    feature_list.append(agent_time_quanta['distance']/1000) #14
    feature_list.append(agent_time_quanta['dotProduct']) #15


    action_list = []
    action_list.append(agent_time_quanta['moveDirection']['x']) #1
    action_list.append(agent_time_quanta['moveDirection']['y']) #2
    action_list.append(agent_time_quanta['lookRotationDelta']['x']) #3
    action_list.append(agent_time_quanta['lookRotationDelta']['y']) #4
    action_list.append(1 if agent_time_quanta['Attack'] else 0) #5
    action_list.append(1 if agent_time_quanta['Reload'] else 0) #6
    action_list.append(1 if agent_time_quanta['thrust'] else 0) #7
    action_list.append(1 if agent_time_quanta['crouch'] else 0) #8
    action_list.append(1 if agent_time_quanta['sprint'] else 0) #9
    action_list.append(1 if agent_time_quanta['slide'] else 0) #10

    return feature_list, action_list

all_agent_data = []
for filename in json_files:
    file_path = os.path.join(folder_path, filename)
    with open(file_path, 'r') as f:
        data = json.load(f)
    agent_features = {}
    agent_features['agent_id'] = filename.split('.')[0]
    agent_features['features'] = []
    for i in range(len(data)):
        feature_list, action_list = prepare_features(data[i])
        agent_features['features'].append((feature_list, action_list))

    # Create the directory if it doesn't exist
    output_dir = "/Users/vaibhavmishra/Desktop/btx-game-aicode/clash_squad_agent_partitioned_features"
    
    # Save agent_features as JSON file
    output_file = os.path.join(output_dir, filename)
    with open(output_file, 'w') as f:
        json.dump(agent_features, f, indent=2)
    del agent_features
    import gc
    gc.collect()


# print(all_agent_data)
```
